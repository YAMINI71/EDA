```python
#Importing libraries
import pandas as pd
import numpy as np
#Read csv file
df=pd.read_csv("minorproject.csv", index_col=0)
df.head() #First 5 rows
```

```python
df.shape
```

**Cleaning (fix, replace, drop, etc) of values <br>
  Drop unused columns, Replace Values & Using mean/median of datas to replace empty (0) entries**

```python
#Dropping columns
df=df.drop(["imdb_id","popularity","homepage","director","tagline","keywords","overview","production_companies","vote_count","vote_average","release_date"
],axis=1)
#Replace 0 to NaN
df.replace(0,np.nan, inplace=True) 

#Fixing missing values (of columns of interest) to mean of data
df["budget"]=df["budget"].fillna(np.mean(df["budget"]))
df["revenue"]=df["revenue"].fillna(np.mean(df["revenue"]))

df.head()
```

# 2) What is the average number of words in movie titles between the years 2000-2005?

```python
df=pd.read_csv("minorproject.csv", index_col=18) #here index col is year of release
sort3=df.loc[[2000,2001,2002,2003,2004,2005],["original_title"]] #rows of interest and column of interest
count=0 #initialise count
sort3 #show df
#Here there are 1678 rows ie 1678 is the total number of movies
```

```python
for item in sort3.original_title:
    count=count+len((item.split())) #split wherever whitespace, add item to count
 #Here, Count is the Total number of words
print("Total number of words in movie titles in the given years=", count)
print("The average number of words in movie titles between the years 2000 to 2005=",count/1687, "words.")
```

# 3) What is the most common Genre for Vin Diesel & Emma Watson movies?

**3.a) Vin Diesel**

```python
df=pd.read_csv("minorproject.csv", index_col=13) #index col is genre
df.fillna(value="False", inplace=True) #fill blanks with False
df1 = df['cast'].str.contains("Vin Diesel") #True wherever cast contains Vin Diesel

df1=df1.to_frame() #to convert series to df
df1=df1[df1["cast"]==True] #selecting rows(genres) where cast (Vin Diesel) is there (True)

df1.groupby("genres")["cast"].sum() #groups by "genre", takes sum of unique genres in df1 (where df1 contains genres where Vin Diesel is present)
```

**Most Common Genre for Vin Diesel= Action, Action|Crime|Thriller, Action|Drama,Action|Thriller|Crime**
                         

**3b) Emma Watson**

```python
#Similarly for Emma Watson
df=pd.read_csv("minorproject.csv", index_col=13)
df.fillna(value="False", inplace=True)
df1 = df['cast'].str.contains("Emma Watson")

df1=df1.to_frame() #to convert series to df
df1=df1[df1["cast"]==True] #selecting rows(genres) where cast (Emma Watson) is there (True)

df1.groupby("genres")["cast"].sum()
```

**Most Common Genre for Emma Watson -  Adventure|Fantasy|Family(6)**

# 4) Which are the movies with the most and least earned revenue?

```python
df=pd.read_csv("minorproject.csv", index_col=0) #Read CSV
df.replace(0,np.nan, inplace=True) #Replace 0 values with na

df["revenue"]=df["revenue"].fillna(np.mean(df["revenue"])) #Replace na values with mean of Revenue
sort4=df.sort_values("revenue", ascending=True) #Sort in ascending to find least earned revenue
sort4.head(1)
```

```python
sort5=df.sort_values("revenue", ascending=False)
sort5.head(1)
```

**Movie with most earned revenue= Avatar** <p>	
**Movie with least earned revenue= Shattered Glass**

# 5) What is the average runtime of movies in the year 2006?


```python
df=pd.read_csv("minorproject.csv", index_col=-3) #index is the release year
rt=df.loc[2006, "runtime"] #Selecting row 2006 and column of interest (runtime)
rt=rt.to_frame() #Series to DF
rt["runtime"].mean()
```

**The average runtime of movies in the year 2006=101.68 minutes**
